# Games of incomplete information
---

<br>

### Motivation

Everything we have done in the part 1 for static or dynamic game, perfect or imperfect information (where imperfect is related to simulatenous moves and chance) was based on the assumption that:

1. All players know the game, including the possible actions and payoffs of all other players
2. All players know that other players know and so on recursively (**knowledge is common knowledge**)

Realistically, we never quite know the payoff of other players. For instance, we might know the amount of money they might get from a deal, but not how much this money is really worth for them (the utility of money is very rarely the identity function).

<br>

### Incomplete information

Everything that players know about themselves but that other players do not know about them is encapsulated in what we call the **type of a player**. We will assume that players know the joint distribution of probability of types of players (**prior knowledge**) and that this knowledge is common knowledge.

A **incomplete information game is a game in which players have types**. In such games, strategies of each players have to include their own type as argument, and the value function of each player depend on the type of the player and might also depend on the types of other players as well.

<br>

### Motivational example

In the market entry game, the only sub-game perfect equilibrium is for the player 1 to enter the market, and for the player 2 to accept this entry. This is what a pair of *rational* player would do.

But **rational is relative to the payoffs**, the values that each player place on each outcome. There is no judgement whatsoever on how these payoffs are evaluated and they can be something else than money.

For some types of player 2, the outcome of fight is -1/2 instead of -1/-1, because this player is not so much interested about money than feeling unique.

If player 1 is uncertain about the **intrinsic motives** of player 2, then we can introduce types to differentiate between these behaviors.

# Static Bayesian Game of Incomplete information
---

A static bayesian game of incomplete information is the equivalent with types of the static game with moves that can be either simultaneous or not.

<br>

### Definition

A game of *N* player is composed of the following:

* A set of actions $\{A\}_n$ for each of the *N* players
* A set of types $\{\Theta\}_n$ for each of the *N* players
* A set of probabilities $\{\phi\}_n = p(\theta_{-i}|\theta_i)$ where $\theta_i$ is the player type and $\theta_{-i}$ is the opponents' types
* Value functions $\{v\}_n$ for each player (which takes tall types as input)

The games unfolds in the following order: each player gets its type and gets to know its own type, then all players pick an action, and then the payoffs are realized by calling the function $v_i$ for each player *i*.

<br>

### Bayesian Nash equilibrium

A **Bayesian Nash equilibrium** is a set of strategies, one for each player, such that none of the player, whatever is type, has any benefit in deviating from it. It is the same concept as a Nash equilibrium, but for the fact that each player has to formulate one strategy for each of its types, and that this strategy must be the **best response in expectation** for the distribution of types of opponents:

&emsp; $\displaystyle \mathbb{E}_{\theta_{-i}}[v_i(s, \theta_i)] \ge \mathbb{E}_{\theta_{-i}}[v_i(s', \theta_i)]$
&emsp; where *s'* is *s* with a different move for player *i*

Note that if the strategy of player *i* is the best response for each of its type (highest payoff with opponents' actions fixed), the expectation of its payoff is at its highest point as well:

&emsp; $\displaystyle \mathbb{E}_{\Theta}[v_i(s, \theta_i)] \ge \mathbb{E}_{\Theta}[v_i(s', \theta_i)]$
&emsp; where *s'* is *s* with a different move for player *i*

So it turns out that to find the Nash equilibrium, **we can solve an incomplete game with the same tools as an imperfect game**: we do not need one normal form matrix for each player $i$, and each type $\theta_i$, we can still use a single normal form matrix and look for deviation of individual player best responses:

1. For player $j \ne i$, it appears that player $i$ is playing a mixed strategy
2. For player $i$, maximizing its "mixed strategy" is equivalent to maximzing all types

<br>

  </tr>

* In the extreme case, if a player is twice as strong as the other, its payoff is 0.5
* In the opposite case, if a player is half as strong as the other, its payoff is -1

As a starter, we can compute the average payoff for the player 1 given its strength, against an opponent that is always willing to fight:

&emsp; $\displaystyle v_1(F, \theta_1) = \mathbb{E}\big[1-\frac{\theta_2}{\theta_1}\big] = 1 - \frac{1}{\theta_1} \int_1^2 \theta_2 d\theta_2 = 1 - \frac{1.5}{\theta_1}$

&emsp; $v_1(F, \theta_1) \ge 0 \iff \theta_1 \ge 1.5$

Above the threshold of strength 1.5, **F** is strictly dominant over **B**, so the player should always choose to fight. But the situation is symetric, and the opponent will not always want to fight. We note **T** the common threshold at which a player is ready to fight and compute the average payoff of fighting:

&emsp; $\displaystyle v_1(F, \theta_1) = 1 - \frac{1}{\theta_1} \int_T^2 \theta_2 d\theta_2 = 1 - \frac{2}{\theta_1} + \frac{T^2}{2 \theta_1}$

&emsp; $\displaystyle v_1(F, \theta_1) \ge 0 \iff \theta_1 \ge 2 - \frac{T^2}{2} \iff T = 2 - \frac{T^2}{2} \iff T^2 + 2 T = 4$

&emsp; $T^2 + 2 T = 4 \implies (T+1)^2 = 5 \implies T = \sqrt{5} - 1 \simeq 1.23$

And so we conclude that players should always fight if $\theta \ge 1.23$. In case we re-establish the positive payoffs in case of the other backing down like below:

# Mechanism design
---

Reverse game theory: we want to build a game so that the players that will participate in it will go to a chosen outcome.

<br>

### Motivation

We have a set of possible outcomes *X* and a group of player, each with its type $\theta_i$ and preferences regarding these possible outcomes. The utility of each player for the outcome *x* is $u_i(x, \theta_i)$.

We call **decision rule** $x(\theta)$ the outcome the game designer desires. This outcome generally depends on the types of all the players. The goal of the game designer is to design a game in which this decision rule is among the Bayesian Nash equilibrium of the game, or even better, a dominant strategy of this game.

To do so, the game designer can design a game in which the payoff of the outcome is modifed by a **transfer rule** $m_i(\theta)$ which assigns for each player a positive or negative transfer of value, resulting in the payoff:

&emsp; $v_i(x, \theta_i) = u_i(x, \theta_i) + m_i(x, \theta_i)$
&emsp; such that
&emsp; $\sum_i m_i(x, \theta) \le 0$

These transfers represent transfer of values between players (for instance money, that comes in the form of taxes or state aides) such that the balance stays positive (the game designer can keep part of the surplus, this allows to design some games such as auctions).

These transfers are meant to **incentivise to do the right thing by aligning the personnal incentives with the social incentives**: when player aim at maximizing their own payoff, they naturally aim at implementing the decision rule desired by the mechanism designer.

<br>

### Definition: Mechanism

The combination of a decision rule $x(\theta)$ with the transfers rule $m_i(\theta)$ is called a **choice rule**. The game designer choses this choice rule to push the players toward the desired decision rule.

&emsp; $f(\theta) = (x(\theta), m_1(\theta) \dots m_N(\theta)) : \Theta \longmapsto Y$

A mechanism $\Gamma$ is a collection of *N* actions sets (one for each player) together with an outcome function *g* that maps from the strategy to an outcome and a set of transfers:

&emsp; $\Gamma = \langle A_i, g \rangle$
&emsp; where
&emsp; $g: S_1 \times \dots \times S_N \longmapsto Y$

A mechanism $\Gamma$ **implements the choice rule** $f(\theta)$ if there is a Bayesian Nash equilibrium $(s_1(\theta_1), \dots s_N(\theta_N))$ such that the outcome of these strategies is equal to the choice rule for all possible values of $\theta$:

&emsp; $g(s_1(\theta_1), \dots s_N(\theta_N)) = f(\theta)$

What matters for the game designer is to **find such a choice rule (design the mechanism) so that the decision rule** desired by the game designer is part of a Bayesian Nash equilibrium.

<br>

### Relevation principle

In an equilibrium, the player knows that the mechanism implements a choice rule *f* and choose to stick to it because they know that it is in their own advantages (thanks to the transfer rule that **align the personnal incentives with the social incentives**). For them, it is equivalent to tell their types and let the mechanism designer implement the choice rule direction. 

The revelation principle says that if a decision rule can be implemented by the mechanism designer using some mechanism, it can be implemented by a **direct revelation game** in which all players anounce their true type and the mechanism designer uses his decision rule.

This revelation game should incentivise the players to give their true type in their own benefit. If such a game cannot implement the decision rule desired by the game designer, then no other mechanism can.

<br>

### Dominant strategy equilibrium

If a mechanism implements a choice rule, it means there is a Bayesian Nash Equilibrium whose outcome for all types of players matches the decision rule. If the player belives the other player will play this Nash equilibrium, they have no incentives to derive from it.

A **dominant strategy equilibrium** is a stronger version of the Nash equilibrium because it relieves the players to form beliefs about opponent, nor about their types and their distribution: such an equilibrium is the best response to any strategy that any opponent could form.

We say that a **choice rule is implementable in dominant strategies** if the outcomes decided by the decision rule correspond to the outcomes of the dominant strategy equilibrium.

<br>

### Pareto optimality

We say that a decision rule *x* is pareto optimal if it maximises the sum of the utility of each player for each possible combination of types $\theta$ of the players:

&emsp; $\displaystyle x(\theta) \in \underset{x \in X}{\operatorname{argmax}} \sum_i u_i(x, \theta_i)$

We can make sure that the Pareto optimal is also the dominant strategy equilibrium by designing a transfer value mechanism such that the loss in utility of a player by moving from strategy *s* of the Pareto optimal to strategy *s'* are more than compensated by the tranfers: this is always possible because by definition the sum of utility of *s* is higher than the one of *s'*.

### Example of the two brothers

Two brothers wants to travel abroad. There are three possible touristic destinations (A, B and C) and each brother has his preferences on the destination:

* Brother 1 (type $\theta_1$) prefers A (3) to B (2) to C (1)
* Brother 2 (type $\theta_2'$) prefers C (3) to B (2) to A (1)
* Brother 2 (type $\theta_2''$) prefers B (3) to A (2) to C (1)

The parents wants to implement the following decision rule:

&emsp; $f(\theta_1, \theta_2') = B$
&emsp; and
&emsp; $f(\theta_1, \theta_2'') = A$

For that to work, the parents must manage brother 2 to anounce his real type, but brother 2 will not always give his real type because if he is of type $\theta_2''$, it is in his best interest to lie, so that the destination is B. To incentivise him to do the right thing, we can use the **pivotal mechanism** to add the following transfer rules:

&emsp; $m_2(\theta_2') = 2 - 3 = -1$ (inclusion of brother 2 would change the outcome of brother 1 to worse alternative)

&emsp; $m_2(\theta_2'') = 3 - 3 = 0$ (inclusion of brother 2 would not change the outcome of brother 1)

With that, brother 2 has no interest in lying to be $\theta_2'$ because the added utility is compensated by the transfer in the other direction.
