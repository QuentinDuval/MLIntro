### Dropout

Paper: http://jmlr.org/papers/volume15/srivastava14a/srivastava14a.pdf

A classical method to improve the generalization properties of models is to train $K$ different models (or the same model but on different data set) and combine them into a big model through averaging, an ensemble technique known as **bagging**:

&emsp; $\displaystyle p(t|x) = \sum_{model} p(model|X,T) p_{model}(t|x)$
&emsp; which is a direct application of **marginalizing** as Bayesian methods would

To have a benefit, bagging must apply to models that are as independent to each other as possible. This means that the models must have different architecture, or be trained on different data set, so that their error are not related to each other. The idea behind Dropout is to find a **cheap way** (at training but also at runtime) to bag neural networks with **different architectures**:

* **During training, units are dropped** with a probability $p$ which is an hyperparameter (but 0.5 usually works best)
* **During testing, units outputs are scaled by** $p$, so that the output corresponds to their expected value

During the training phase, the dropout is applied independently for each mini-batch. This is equivalent to train an exponential number of different **thin architectures** ($N$ units means $2^N$ combinations), one by mini-batch, with the different architectures sharing their parameters. During the testing phase, the scaling is equivalent to taking the average of the output of these networks.

Technically, the error function is it optimizing the geometric mean of an exponential number of models:

&emsp; $\displaystyle p(t|x) = \Big ( \prod_k p_k(t|x) \Big )^\frac{1}{K} \iff \log p(\mathcal{D}) = \sum_n \frac{1}{K} \sum_k \log p_k(t_n|x_n)$
&emsp; the log likelihood of these $K$ thin networks

**Why does it work?** By breaking a huge number of interconnections, units are encouraged to **limit fragile complex cooperation scheme** that might improve performance on the training data:

* **simplicity** is encouraged, leading to more robustness to change and overfitting to training data
* units are encouraged to learn more stable functions (inputs to hiddent layers gets **noisy**)
* the overall architecture favors **redundancy** (paths are cut), leading to several experts agreing on the output
* breaking co-adaptation means each sub-net is **optimized for cooperation** rather than individual fitness
* breaking complex schemes encourages each unit to be **self-sufficient** and do something useful on its own

**Note:** The paper makes reference to the *sexual reproduction* and Generic Algorithms. The mutation part of GA is like gradient descent, while the crossing part allows to escape local minimum, but also favors simplicity, because it breaks the ability to build complex interactions schemes (a given version of a gene cannot depend too much on specific versions of other genes, or else most offsprings would not be viable):

* Versions of genes are therefore optimized for cooperation rather than for complex inter-relation with specific genes
* This cooperation makes possible large changes, allowing to escape local minimum

**The shape of the model influence the kind of solution the optimizer will select**. Dropout pushes the optimizer toward regions of the solution space with less agressive performance on the training set but with other qualities that make the model more adaptable to change in context, more robust, and less overfitting.

### Batch normalization

Paper: https://arxiv.org/pdf/1502.03167.pdf

An ANN will learn faster and work better if the training inputs and test inputs follow the same input distribution (i.e. are on the same manifold). Since a multi-layer ANN can be seen as several layers on top of each other, the same principle should be true for any hidden layer.

But in practice, changes of the parameters of the lower layers have a big impact on the inputs of the higher layers. Therefore the higher layers have to adapt to change of their input distribution constantly, and this significantly slows down training. The idea behing batch normalization is that, **training is made easier if the distribution of inputs of each layer is made stable by normalizing its inputs**.

The approach is normalize each dimension independently, ignoring possible covariance between dimensions:

&emsp; $\displaystyle \tilde{x} = \frac{x - E[x]}{\sqrt{\sigma^2[x]}}$
&emsp; with
&emsp; $\displaystyle E[x] = \frac{1}{m} \sum x_i$ the sample mean
&emsp; $\displaystyle \sigma^2[x] = \frac{1}{m-1} \sum (x_i - E[x])^2$ the sample variance

**First problem**: the mean and variance can significantly change between batches, especially for higher layers. We therefore keep some momemtum for the mean and variance estimations (running mean and running variance, which are variants of gradient updates):

&emsp; $\displaystyle \mu_{new} \leftarrow \mu_{old} + \alpha \Big ( \frac{1}{m} \sum x_i - \mu_{old} \Big )$
&emsp; and
&emsp; $\displaystyle \sigma_{new}^2 \leftarrow \sigma_{old}^2 + \alpha \Big ( \frac{1}{m-1} \sum (x_i - \mu_{new})^2 - \sigma_{old}^2 \Big )$

**Second problem**: if we keep the values centered all the time, we lose the ability to exploit the non-linear regime of non-linearities such as sigmoids. We therefore associates two learnable parameters by dimensions $\gamma$ and $\beta$ to scale the inputs:

&emsp; $\displaystyle \hat{x} = \gamma \frac{x - E[x]}{\sqrt{\sigma^2[x]}} + \beta$

**Last problem**: the numerical stability of the division by the variance, which require introducing a small $\epsilon$ under the square root:

&emsp; $\displaystyle \hat{x} = \gamma \frac{x - E[x]}{\sqrt{\sigma^2[x] + \epsilon}} + \beta$

**Resulting properties of Batch Normalization**:

* Accelarates training, by allowing each layer to keep the same input distribution as input
* Make the layer insensitive to parameter scaling in lower layers $\text{BN}(Wx) = \text{BN}(\alpha W u)$
* Make the layer insensitive to explosing gradients due to scaling $\nabla_x \text{BN}(Wx) = \nabla_x \text{BN}(\alpha W u)$
* Offer limited **regularization** as each dimensions gets links with the other via the mean and variance

**Guidelines with Linear and CNN layers**:

* Batch Normalization is better applied **before any non-linearity**
* **Linear layers do not need bias** when followed by a Batch Normalization (because it includes a bias)
* CNN requires having **weight sharing** between the activations (same as for CNN)

```python
class BatchNorm1d(nn.Module):
    def __init__(self, num_features: int, epsilon: float = 1e-5, momentum: float = 0.1):
        super().__init__()
        self.num_features = num_features
        self.momentum = momentum
        self.epsilon = epsilon
        self.weight = nn.Parameter(torch.FloatTensor(num_features)) # one weight by dimension
        self.bias = nn.Parameter(torch.FloatTensor(num_features))   # one bias by dimension
        self.register_buffer('running_mean', torch.zeros(num_features)) # register_buffer makes not visible for optimizer
        self.register_buffer('running_var', torch.ones(num_features))   # register_buffer makes not visible for optimizer
        nn.init.ones_(self.weight)
        nn.init.zeros_(self.bias)
    
    def forward(self, xs):
        if self.training:
            sample_mean = torch.mean(xs)
            sample_var = torch.mean(xs**2) - sample_mean**2 # TODO - correct with m / (m-1)
            prev_running_mean = self.running_mean.detach() # detach to limit size of graph
            prev_running_var = self.running_var.detach()   # detach to limit size of graph
            self.running_mean = prev_running_mean + (1 - self.momentum) * (sample_mean - prev_running_mean)
            self.running_var = prev_running_var + (1 - self.momentum) * (sample_var - prev_running_var)
            return self.weight * (xs - self.running_mean) / torch.sqrt(self.running_var + self.epsilon) + self.bias
        else:
            prev_running_mean = self.running_mean.detach()
            prev_running_var = self.running_var.detach()
            return self.weight * (xs - prev_running_mean) / torch.sqrt(prev_running_var + self.epsilon) + self.bias


bn = BatchNorm1d(num_features=5)

bn.train()
y = bn(torch.FloatTensor([1., 2., -1, 0., 5]))
print(y)
torch.sum(y).backward()

bn.eval()
y = bn(torch.FloatTensor([5., 5., 5., 5., 5.]))
print(y)
torch.sum(y).backward()
```

### Residual Net

Paper: https://arxiv.org/pdf/1512.03385.pdf

Addresses the **degration in training and testing error that occurs when increasing the depth** of neural networks. Stacking layers should in theory always produce a better solution, because the higher layer should in principle be able to learn the identity function.

But in practice, stacking more layers might lead to degraded training error. This is **not due to overfitting** because overfitting would only affect the testing error. This phenomenon also occurs if we have Batch Normalization layers in between. So this is not due to instability of inputs of layers. The degradation indicates that not all systems are similarly easy to optimize.

Residual networks help training by introducing **shortcut between layers**. These layers allow easier optimization of the network at increasing depth, and allow to avoid the degradation phenomena described above:

* training of networks at equal depth becomes faster in many cases
* training of networks at higher depths becomes feasible (training error keeps decreasing)
* this leads to increased performance in many case, by allowing to use deeper networks

**Why residual?** In Maths, when solving the equation $y = f(x)$ for $x$, we call residual the difference between the output of our current best solution $x_0$ with the target: $y - f(x_0)$. Similarly, when solving $y = f(x)$ for $f$, we call *residual* the difference between our best solution $f_0$ and $f$. Techniques of optimization based on optimizing the residuals are numerous and effective.

In our case, say we want to learn a function $\mathcal{H}(x)$, we could alternatively try to learn the difference between the input $x$ and desired output $\mathcal{H}(x)$, that is, we would try to learn the residual between the identity function and the target function.

So instead of learning $\mathcal{H}(x)$, we try to learn $\mathcal{F}(x) = \Delta(x) = \mathcal{H}(x) - x$, and add this to the input $x$, to recover the function $\mathcal{H}(x) = x + \mathcal{F}(x)$. The skipped layers represent the function $\mathcal{F}$ we are trying to learn.

**Why does it work?** We reasonned before that additional layer should in theory not reduce the training error because they could simply learn the *identity* function. The shortcuts allows this to happen practically, because using the shortcut means using the identity mapping. They may also have an effect on *vanishing/exploding* gradients.

> *The degradation problem suggests that the solvers might have difficulties in approximating identity mappings
by multiple nonlinear layers. With the residual learning reformulation, if identity mappings are optimal, the solvers
may simply drive the weights of the multiple nonlinear layers toward zero to approach identity mappings.*

**Variants:** Shortcuts can skip over one or several layers. The shortcut can be something more complex than the identity function. In which case we try to learn the residual between 2 functions. The paper argues that this does not lead to better results, while incuring computational cost (while the residual with the identity is very cheap).

**Notes:** Skipping a single layer does not really bring advantages. If applied before the non-linearity, it is equivalent to initializing with the identity matrix. If applied the non-linearity, it does not either, since most non-linearity are ReLU() nowadays.

# Example: Learning a sinusoid
---

The function to learn, with very few points:

```python
train_xs = np.random.uniform(-13, 13, size=40)
train_ys = np.sin(train_xs)

test_xs = np.arange(-15, 15, 0.1)
test_ys = np.sin(test_xs)

plt.figure(figsize=(15,6))
plt.scatter(train_xs, train_ys, color='green')
plt.scatter(train_xs, train_ys, s=200, color="lightgreen", alpha=0.3)
plt.plot(test_xs, test_ys, color='orange')
plt.ylim(-1.5, 1.5)
plt.show()
```

<br>

### Vanilla Neural Net

Note that this is pretty amazing that an ANN with so many weights can learn a function with so few points.

```python
class VanillaFullyConnectedNN(nn.Module):
    def __init__(self, hidden_levels: int, hidden_size: int):
        super().__init__()
        levels = []
        for _ in range(hidden_levels):
            levels.append(nn.Linear(hidden_size, hidden_size))
            levels.append(nn.ReLU())
        self.fc = nn.Sequential(
            nn.Linear(1, hidden_size),
            nn.ReLU(),
            *levels,
            nn.Linear(hidden_size, 1))
    
    def forward(self, xs):
        return self.fc(xs)

    
def train(model, train_xs, train_ys, epoch_nb: int, lr: float):
    model.train()
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    ds = data.TensorDataset(torch.FloatTensor(train_xs).unsqueeze(-1),
                            torch.FloatTensor(train_ys).unsqueeze(-1))
    loader = data.DataLoader(ds, batch_size=100, shuffle=True)

    losses = []
    for _ in range(epoch_nb):
        episode_loss = 0.
        for x, y in loader:
            y_pred = model(x)
            loss = criterion(y_pred, y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            episode_loss += loss.item()
        losses.append(episode_loss)
    return losses


def predict(model, test_xs):
    model.eval()
    return model(torch.FloatTensor(test_xs).unsqueeze(-1)).detach().numpy()


model = VanillaFullyConnectedNN(hidden_levels=5, hidden_size=256)
losses = train(model, train_xs, train_ys, epoch_nb=1000, lr=2e-4)
ys_pred = predict(model, test_xs)
print('Final loss', losses[-1])
plt.plot(losses)
plt.ylim(0, 1.0)
plt.show()

plt.figure(figsize=(15,6))
plt.scatter(train_xs, train_ys, color='green')
plt.plot(test_xs, test_ys)
plt.plot(test_xs, ys_pred, color='orange')
plt.ylim(-1.5, 1.5)
plt.show()
```

### Residual Networks

```python
class ResLinearBlock(nn.Module):
    def __init__(self, input_size: int, hidden_size: int, output_size: int):
        super().__init__()
        self.l1 = nn.Linear(input_size, hidden_size)
        self.l2 = nn.Linear(hidden_size, output_size)
        self.shortcut = nn.Linear(input_size, output_size)
        self.relu = nn.ReLU()
    
    def forward(self, xs):
        ys = self.l1(xs)
        ys = self.relu(ys)
        ys = self.l2(ys)
        xs = self.shortcut(xs)
        return self.relu(xs + ys)


class ResidualNet(nn.Module):
    def __init__(self, hidden_size: int):
        super().__init__()
        self.fc1 = nn.Linear(1, hidden_size)
        self.l1 = ResLinearBlock(hidden_size, hidden_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.l2 = ResLinearBlock(hidden_size, hidden_size, hidden_size)
        self.fcN = nn.Linear(hidden_size, 1)
        self.relu = nn.ReLU()
    
    def forward(self, xs):
        xs = self.relu(self.fc1(xs))
        xs = self.l1(xs)
        xs = self.relu(self.fc2(xs))
        xs = self.l2(xs)
        return self.fcN(xs)
        

model = ResidualNet(hidden_size=256)
losses = train(model, train_xs, train_ys, epoch_nb=1000, lr=2e-4)
ys_pred = predict(model, test_xs)
print('Final loss', losses[-1])
plt.plot(losses)
plt.ylim(0, 1.0)
plt.show()


plt.figure(figsize=(15,6))
plt.scatter(train_xs, train_ys, color='green')
plt.plot(test_xs, test_ys)
plt.plot(test_xs, ys_pred, color='orange')
plt.ylim(-1.5, 1.5)
plt.show()
```

### Batch Normalization

Increases the speed of learning tremendously! It even goes to overfitting.

```python
class ResidualNetWithBN(nn.Module):
    def __init__(self, hidden_size: int):
        super().__init__()
        self.fc1 = nn.Linear(1, hidden_size)
        self.l1 = ResLinearBlock(hidden_size, hidden_size, hidden_size)
        self.bn1 = BatchNorm1d(hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.l2 = ResLinearBlock(hidden_size, hidden_size, hidden_size)
        self.bn2 = BatchNorm1d(hidden_size)
        self.fcN = nn.Linear(hidden_size, 1)
        self.relu = nn.ReLU()
    
    def forward(self, xs):
        xs = self.relu(self.fc1(xs))
        xs = self.l1(xs)
        xs = self.bn1(xs)
        xs = self.relu(self.fc2(xs))
        xs = self.l2(xs)
        xs = self.bn2(xs)
        return self.fcN(xs)
        

model = ResidualNetWithBN(hidden_size=256)
losses = train(model, train_xs, train_ys, epoch_nb=1000, lr=2e-4)
ys_pred = predict(model, test_xs)
print('Final loss', losses[-1])
plt.plot(losses)
plt.ylim(0, 1.0)
plt.show()


plt.figure(figsize=(15,6))
plt.scatter(train_xs, train_ys, color='green')
plt.plot(test_xs, test_ys)
plt.plot(test_xs, ys_pred, color='orange')
plt.ylim(-1.5, 1.5)
plt.show()
```
